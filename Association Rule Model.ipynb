```python
import csv
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
dataset = []
```

```python
with open('CraftZoneStoreDataSet.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)
    for row in csv_reader:
        dataset.append(row)
```

```python
data = [[item for item in sublist if item != ''] for sublist in dataset]
```

```python
data[:5]
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
df1 = apriori(df,min_support=0.05,use_colnames=True)
df1
```

```python
df1.sort_values(by="support",ascending=False)
```

```python
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
df1
```

```python
df1[(df1['length']==2) & (df1['support']>=0.05)]
```
